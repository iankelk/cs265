```python
# Conduct a search of node sizes from 1 to 16
node_sizes = range(1,17)
time_list = []

for ns in node_sizes:
    res = !./benchmark -n {ns}
    x = re.search(r"<(.*)>", res[2])
    time_list.extend([float(x.group(1))])
```

```python
plt.figure(figsize=(16,6))
plt.plot( range(1,17), time_list)
plt.xlabel("Node Size", fontsize=14)
plt.ylabel("Seconds to Complete", fontsize=15)
plt.title("Seconds to Complete vs Node Size", fontsize=15)
plt.grid(True)
```

```python
# Conduct a search of q values from 1 to 10
q_values = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]
time_list = []

for q in q_values:
    res = !./benchmark -n 4 -s 49999 -q {q}
    x = re.search(r"<(.*)>", res[2])
    time_list.extend([float(x.group(1))])
```

```python
plt.figure(figsize=(16,6))
plt.plot( q_values, time_list)
plt.xlabel("Q Value", fontsize=14)
plt.ylabel("Seconds to Complete", fontsize=15)
plt.title("Seconds to Complete vs Q Value", fontsize=15)
plt.grid(True)
```
